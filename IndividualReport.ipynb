Predicting Newsletter Subscription from player demographics and behaviour

```r
library(tidyverse)
```

```r
library(dplyr) # Assuming you are using 'dplyr' with the pipe
# Calculate the mean age
players<-read_csv("data/players.csv")
sessions<-read_csv("data/sessions.csv")
mean_stats <- players |>
  summarise(
    mean_age = mean(as.numeric(Age), na.rm = TRUE),
    mean_played_hrs = mean(as.numeric(played_hours), na.rm = TRUE)
  )

mean_stats
```

1) Dataset

Players - 196 observations, 7 variables
Sessions - 1555 observations, 5 variables

Players Summary 
- experience (chr): experience of a player ranging from Beginner, Regular, Amateur, Veteran and Pro
- subscribe (lgl): whether player is subscribed to newsletter or not
- hashedemail (chr): hashed email of player
- played_hours (dbl): hours played by player
- name (chr): name of player
- gender (chr): gender of player
- Age (dbl): age of player
- mean of Age - 21.14
- mean of Hours Played - 5.85

Sessions Summary
- hashedemail (chr) - same as in Players dataframe
- start_time (chr) - start time of playing
- end_time (chr) - end time of playing
- original_start_time (dbl) - same as above in number format
- original_end_time (dbl) - same as above in number format

Predicting Newsletter Subscription from player demographics and behaviour

```r
library(tidyverse)
library(dplyr)
library(repr)
library(GGally)
options(repr.matrix.max.rows = 6)
```

```r
players<-read_csv("data/players.csv")
sessions<-read_csv("data/sessions.csv")
mean_stats <- players |>
  summarise(
    mean_age = mean(as.numeric(Age), na.rm = TRUE),
    mean_played_hrs = mean(as.numeric(played_hours), na.rm = TRUE)
  )

mean_stats
```

1) Dataset

Players - 196 observations, 7 variables
Sessions - 1555 observations, 5 variables

Players Summary 
- experience (chr): experience of a player ranging from Beginner, Regular, Amateur, Veteran and Pro
- subscribe (lgl): whether player is subscribed to newsletter or not
- hashedemail (chr): hashed email of player
- played_hours (dbl): hours played by player
- name (chr): name of player
- gender (chr): gender of player
- Age (dbl): age of player
- mean of Age - 21.14
- mean of Hours Played - 5.85

Sessions Summary
- hashedemail (chr) - same as in Players dataframe
- start_time (chr) - start time of playing
- end_time (chr) - end time of playing
- original_start_time (dbl) - same as above in number format
- original_end_time (dbl) - same as above in number format

Issues I see:

- Name is visible which could be a confidentiality issue

- Duplicates


2)
Broad question: What player characteristics and behaviours predict subscribing to the game-related newsletter?

Specific question: Can the hours played by a player and age of a player predict whether a player subscribes to the newsletter?

By using the dataset of players, we can analyse and see whether there is a pattern seen in the hours played and age that results in whether a player will subscribe to the newsletter. At the moment, this prediction is not very complicated, hence,it does not require much data wrangling.

```r
3. 
specific<- players |>
select(subscribe, played_hours, Age)

specific

options(repr.plot.height = 8, repr.plot.width = 7)
ageplot <- specific |> 
     ggplot(aes(x= subscribe, y = Age)) +
geom_point()
playplot <- specific |> 
     ggplot(aes(x= subscribe, y = played_hours)) +
geom_point()
ageplot
playplot
```
